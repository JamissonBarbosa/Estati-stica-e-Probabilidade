
1. Querendo determinar o peso médio de nicotina dos cigarros de sua produção,
um fabrticante recolheu uma amostra de 25 cigarros, obtendo x = 38 e
s = 0, 5. Teste se o peso médio de nicotina é inferior a 40 mg ao nı́vel de
5%.

Respostas:

$s = 0,5$ 

$\alpha = 0,05$

n = 25  $\phi$ = 24

$\overline{x} = 38$

$\mu = 40$

$H_0: \mu < 40$

$H_1: \mu \geq 40 $

$T_{\phi,\alpha}$ = $T_{24, 0.025}$ = 2,064

$T_{calc} = \frac{\overline{\mathit{x}} -\mu}{\frac{\sigma }{\sqrt{n}}} = \frac{38-40}{\frac{0,5}{\sqrt{25}}} = -20$

$\alpha = 5\%, t = 2,064$

RNR = [-2.064, 2.064]

com T_{calc} está na regiao critica, rejeitamos H_0
